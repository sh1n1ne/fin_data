```julia
densitykm2_edge = 0:300:100000 
df.densitykm2_bin = cut(df[!,:"density (km²)"], densitykm2_edge)
combine(groupby(df, :densitykm2_bin), nrow)
```

```julia
!Pkg.add("Plots")
```

```julia
using Plots
b_range = range(0, 300, length=10)
#0代表區間範圍的最小值、1000代表區間範圍的最大值、10為間距，請依照自己的資料調整
histogram(df[!,:"density (km²)"],bins=b_range)
xlims!(0, 300)
```

```julia
histogram(df[!,:"density (km²)"],bins=b_range, normalize=:pdf)
xlims!(0, 300)
```

```julia
population_2023_edge = 0:1000000:10000000000
df.population_2023_bin = cut(df[!,:"2023 population"], population_2023_edge)
combine(groupby(df, :population_2023_bin), nrow)
```

```julia
using Plots
b_range = range(0, 1000000, length=6)
#0代表區間範圍的最小值、1000代表區間範圍的最大值、10為間距，請依照自己的資料調整
histogram(df[!,:"2023 population"],bins=b_range)
xlims!(0, 1000000)
```

```julia
histogram(df[!,:"2023 population"],bins=b_range, normalize=:pdf)
xlims!(0, 1000000)
```
